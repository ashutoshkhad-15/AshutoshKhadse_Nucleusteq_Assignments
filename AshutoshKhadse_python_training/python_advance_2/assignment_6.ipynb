```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the employee dataset
data = {
    'Name': ['Rahul', 'Priya', 'Amit', 'Anuj'],
    'Age': [25, 30, 28, 35],
    'Department': ['HR', 'IT', 'Finance', 'IT'],
    'Salary': [30000, 50000, 45000, 60000]
}

df = pd.DataFrame(data)

# Set the visual style for all Seaborn plots
sns.set_theme(style="whitegrid")
print("Dataset ready and Seaborn theme applied!")
```

```python
# Task 1: Create Barplot: Department vs Salary
plt.figure(figsize=(8, 5))
sns.barplot(x="Department", y="Salary", data=df, errorbar=None, hue="Department", legend=False)

plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Salary")

plt.show()
```

```python
# Task 2: Create Boxplot: Salary distribution
plt.figure(figsize=(7, 3))
sns.boxplot(x=df['Salary'], color='skyblue')

plt.title('Distribution of Employee Salaries')
plt.xlabel('Salary')

plt.show()
```

```python
# Task 3: Create Heatmap using correlation between Age & Salary
# First, isolate the numeric columns and calculate the correlation
correlation_matrix = df[['Age', 'Salary']].corr()

plt.figure(figsize=(5, 4))
# annot=True puts the actual correlation numbers inside the colored boxes
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", vmin=-1, vmax=1)

plt.title('Correlation Heatmap: Age vs Salary')

plt.show()
```
