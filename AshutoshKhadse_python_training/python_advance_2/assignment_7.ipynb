```python
import pandas as pd

# Task 1: Create dataset and Load into Pandas
data = {
    'Name': ['Rahul', 'Priya', 'Siri', 'Anuj'],
    'Marks': [70, 80, 90, 60],
    'Hours Studied': [2, 3, 5, 1]
}

df = pd.DataFrame(data)
display(df)
```

```python

# Task 2: Add Performance column (Pass if Marks > 65 else Fail)
df['Performance'] = df['Marks'].apply(lambda x: 'Pass' if x > 65 else 'Fail')

print("Student DataFrame:")
display(df)
```
