```python
import pandas as pd

# Task 1: Create dataset and Load into Pandas
data = {
    'Name': ['Rahul', 'Priya', 'Siri', 'Anuj'],
    'Marks': [70, 80, 90, 60],
    'Hours Studied': [2, 3, 5, 1]
}

df = pd.DataFrame(data)
display(df)
```

```python

# Task 2: Add Performance column (Pass if Marks > 65 else Fail)
df['Performance'] = df['Marks'].apply(lambda x: 'Pass' if x > 65 else 'Fail')

print("Student DataFrame:")
display(df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sorting the dataset by Hours Studied makes the line chart read chronologically
df_sorted = df.sort_values(by='Hours Studied')

# Task 3a: Line Chart: Hours vs Marks
plt.figure(figsize=(6, 4))
plt.plot(df_sorted['Hours Studied'], df_sorted['Marks'], marker='o', color='blue', linewidth=2)
plt.title('Line Chart: Hours Studied vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Task 3b: Scatter Plot: Study vs Marks
plt.figure(figsize=(6, 4))
plt.scatter(df['Hours Studied'], df['Marks'], color='red', s=100)
plt.title('Scatter Plot: Study Hours vs Marks')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Task 4: Use Seaborn Barplot: Performance vs Marks
plt.figure(figsize=(6, 4))
sns.set_theme(style="whitegrid")
sns.barplot(x='Performance', y='Marks', data=df, hue='Performance', palette='Set2', legend=False)

plt.title('Seaborn Barplot: Performance vs Marks')
plt.xlabel('Performance')
plt.ylabel('Marks')
plt.show()
```
