```python
import matplotlib.pyplot as plt

# Data
departments = ['HR', 'IT', 'Finance']
employees = [5, 12, 7]
```

```python
# Task 1: Create Bar Chart
# Creating a Bar Chart to compare discrete categorical groupings
plt.figure(figsize=(8, 5))
plt.bar(departments, employees, color='skyblue', edgecolor='black')

# Adding labels and titles makes the chart understandable without context
plt.title('Employee Count per Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')

# Displaying the plot cleanly
plt.show()
```

```python
# Task 2: Create Line Chart
plt.figure(figsize=(6, 4))
plt.plot(departments, employees, marker='o', color='purple', linestyle='-', linewidth=2)

# Adding labels and title
plt.title('Employee Trend by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
# Task 3: Create Histogram using salaries
salaries = [30000, 40000, 50000, 60000, 45000]

plt.figure(figsize=(6, 4))
plt.hist(salaries, bins=4, color='skyblue', edgecolor='black')

# Adding labels and title
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')

plt.show()
```

```python
# Task 4: Create Scatter Plot (Age vs Salary)
# Using dummy ages to match the 5 salary data points
ages = [25, 30, 35, 40, 28]
salaries = [30000, 40000, 50000, 60000, 45000]

plt.figure(figsize=(6, 4))
plt.scatter(ages, salaries, color='red', s=100) # 's' is the size of the dots

# Adding labels and title
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```
