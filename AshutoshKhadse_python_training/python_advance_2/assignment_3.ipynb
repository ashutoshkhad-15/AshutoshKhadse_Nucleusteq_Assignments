```python
import pandas as pd
import numpy as np

# Create the dataset with missing values
data = {
    'Name': ['Rahul', 'Priya', 'Anuj'],
    'Age': [25, np.nan, 29],
    'Salary': [30000, 40000, np.nan]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
df
```

```python
# Task 1: Detect missing values
# .isna() returns True for missing values, and .sum() counts the True values per column
missing_counts = df.isna().sum()

print("Missing Values Per Column:")
print(missing_counts)

print("\nBoolean map of missing values:")
display(df.isna())
```

```python
# Task 2: Replace missing Age with mean
# Calculate the mean age from the existing valid data
mean_age = df["Age"].mean()
print(f"Calculated Mean Age: {mean_age}\n")

# Use .fillna() to replace only the NaN values in the Age column
df["Age"] = df["Age"].fillna(mean_age)

print("DataFrame after Age imputation:")
display(df)
```

```python
# Task 3: Replace missing Salary with 0
df["Salary"] = df["Salary"].fillna(0)

print("Final Cleaned DataFrame:")
display(df)
```
