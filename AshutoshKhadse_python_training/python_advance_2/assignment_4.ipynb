```python
import pandas as pd

# Recreating the employee dataset from the previous assignment
data = {
    'Name': ['Rahul', 'Priya', 'Amit', 'Anuj'],
    'Age': [25, 30, 28, 35],
    'Department': ['HR', 'IT', 'Finance', 'IT'],
    'Salary': [30000, 50000, 45000, 60000]
}

df = pd.DataFrame(data)

print("Employee DataFrame:")
display(df)
```

```python
# Task 1: Find average salary by department
avg_salary = df.groupby('Department')['Salary'].mean()

print("Average Salary by Department:")
avg_salary
```

```python
# Task 2: Find max salary by department
max_salary = df.groupby('Department')['Salary'].max()

print("Max Salary by Department:")
max_salary
```

```python
# Task 3: Count employees per department
# Using .size() counts the total number of rows in each group
employee_count = df.groupby('Department').size()

print("Employee Count per Department:")
employee_count
```
