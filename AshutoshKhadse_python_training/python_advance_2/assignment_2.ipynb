```python
import pandas as pd

# Creating the raw dataset using a dictionary of lists
data = {
    "Name": ["Rahul", "Priya", "Amit", "Anuj"],
    "Age": [25, 30, 28, 35],
    "Department": ["HR", "IT", "Finance", "IT"],
    "Salary": [30000, 50000, 45000, 60000]
}

# Converting to a Pandas DataFrame
df = pd.DataFrame(data)

print("Employee DataFrame:")
display(df)
```

```python
# Task 2: Show first 2 rows
first_two_rows = df.head(2)

print("First 2 Rows:")
display(first_two_rows)
```

```python
# Task 3: Show summary statistics
summary_stats = df.describe()

print("Summary Statistics:")
display(summary_stats)
```

```python
# Task 4: Display only IT employees
it_employees = df[df["Department"] == "IT"]

print("IT Department Employees:")
display(it_employees)
```

```python
# Task 5: Add new column: Bonus = Salary * 0.10
df["Bonus"] = df["Salary"] * 0.10

print("DataFrame with 10% Bonus Column:")
display(df)
```
